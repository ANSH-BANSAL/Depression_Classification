```python
df
```

```python
phq8_features = [
    'PHQ8_Sleep', 'PHQ8_Tired'
]

X = df[phq8_features].fillna(0).astype(int)
```

```python
def int_to_2bit_binary(val):
    return format(val, '02b')
```

```python
def row_to_binary_vector(row):
    binary_string = ''.join(int_to_2bit_binary(val) for val in row)
    return [int(bit) for bit in binary_string]
```

```python
binary_vectors = X.apply(row_to_binary_vector, axis=1)
```

```python
binary_vectors = binary_vectors.tolist()
```

```python
binary_vectors # basis encoding done
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
X = np.array(binary_vectors)
```

```python
X
```

```python
Y = np.array(df['PHQ8_Binary']).astype(str)
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # train-test split
```

```python
from qiskit.circuit.library import ZZFeatureMap, TwoLocal
from qiskit.utils import algorithm_globals
```

```python
from qiskit_aer.primitives import Sampler
from qiskit_machine_learning.algorithms.classifiers import VQC
```

```python
from qiskit.algorithms.optimizers import SPSA, COBYLA, ADAM
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
algorithm_globals.random_seed = 42
feature_map = ZZFeatureMap(feature_dimension=4)
ansatz = TwoLocal(num_qubits=4, rotation_blocks='ry', entanglement_blocks='cz')
optimizer = COBYLA(maxiter=30)
```

```python
pip install scikit-learn==1.2.2
```

```python
sampler = Sampler()
vqc = VQC(feature_map=feature_map,
          ansatz=ansatz,
          optimizer=optimizer,
          sampler=sampler,
          initial_point=np.random.rand(ansatz.num_parameters))
```

```python
vqc.fit(X_train, y_train)
```

```python
y_pred = vqc.predict(X_test)
acc = accuracy_score(y_test, y_pred)
```

```python
print(f"VQC Test Accuracy: {acc:.4f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap="Blues", values_format='d')
disp.ax_.set_title("Confusion Matrix")
```

```python
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, digits=4))
```
